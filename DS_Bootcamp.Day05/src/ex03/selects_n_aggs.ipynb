```python
import pandas as pd
```

## Load json

```python
df = pd.read_json('../ex02/auto.json', orient='records')
df.set_index('CarNumber', inplace=True)
df.sort_index(inplace=True, ascending=False)
df
```

## Make some selects

### Fines more than 2100

```python
filt = df['Fines'] > 2100
df[filt]
```

### Fines more than 2100 and refund equals 2

```python
filt = (df['Fines'] > 2100) & (df['Refund'] == 2)
df[filt]
```

### Rows with models from list ['Focus', 'Corolla']

```python
models = ['Focus', 'Corolla']
filt = df['Model'].isin(models)
df[filt].sort_values(by='Model')
```

### Rows with numbers from list ['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']

```python
numbers = ['Y7689C197RUS', '92928M178RUS',
           '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']
filt = df.index.isin(numbers)
df[filt]
```

## Make some aggregation with the make and model

```python
makes = df.groupby('Make')
makes_n_models = df.groupby(['Make', 'Model'])
```

### The median fines grouped by the make

```python
makes['Fines'].median()
```

### The median fines grouped by the make and the model

```python
makes_n_models['Fines'].median()
```

### The number of fines grouped by the make and the model

```python
makes_n_models['Fines'].count()
```

### The minimum and the maximum fines grouped by the make and the model

```python
makes_n_models['Fines'].agg(['min', 'max'])
```

### The standard deviation of the fines grouped by the make and the model

```python
makes_n_models['Fines'].std()
```

## Make some aggregation with the car number

```python
car_numbers = df.groupby(df.index)
```

### Top car numbers by fines count

```python
most_fines = car_numbers['Fines'].count().sort_values(ascending=False)
most_fines
```

```python
car_numbers.get_group(most_fines.index[0])
```

### Top car numbers by fines amount


```python
most_fine_cost = car_numbers['Fines'].sum().sort_values(ascending=False)
most_fine_cost
```

```python
car_numbers.get_group(most_fine_cost.index[0])
```

### Is there car numbers connected to different models

```python
filt = car_numbers['Model'].nunique() > 1
bool(filt.sum())
```

```python
car_numbers.filter(lambda x: filt[x.name])
```
