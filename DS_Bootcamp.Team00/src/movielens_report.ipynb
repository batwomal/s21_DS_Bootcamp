## Общие сведения о классе **Links()**

Создав экземпляр класса Links() мы открыли для себя возможность обработку к базы **MovieLens** по movieid, а по movieid обработать содержимое **imdb**. Кстати, IMDb — Internet Movie Database (IMDb, в переводе с англ. — «Интернет-база кинофильмов») (по-русски ИМДб) — крупнейшая в мире база данных и веб-сайт о кинематографе. В базе собрана информация (по состоянию на конец января 2013 года) о более чем 2 млн 436 тыс.

Все dunder методы и аттрибуты любого класса (методы, начинающиеся и заканчивающиеся двойными нижними подчеркиваниями) - это методы, которые не предназначены для непосредственного пользовательского вызова, но их интерпретация вызывается внутри интерпретатора Python для выполнения определенных действий внутреннего использования в самом классе.
С остальными методами и аттрибутами мы можем взаимодействовать, например мы вызвали аттрибут **'base_url_imdb'**

```python
%timeit links.base_url_imdb
links.base_url_imdb
```

и убедились в том, что базовая ссылка действительно обращается к ресурсу IMDB.

Мы также можем посмотреть и путь к файлу с базой MovieLens, который указали ранее:

```python
%timeit links.filepath
links.filepath
```

Дальше мы распарсили сайт imdb с помощью файла links.csv и получили данные по каждому, интересующему нас фильму, передав в метод .get_imdb() список из номеров фильмов - movieid:
<div style="text-align: center;">
  <img src="very_important_folder/8.jpg"/>
</div>

```python
%timeit links.get_imdb([193600, 2, 6, 2467, 903, 2176, 930, 2178])
links.get_imdb([193600, 2, 6, 2467, 903, 2176, 930, 2178])
```

На выходе мы получили **список списков**- это такой список, элемент которого также является списком! В нашем случае первым элементом списка является заголовок, который отражает название позиции аргумента в последующих списках, а последующие списки - это информация по каждому отдельному фильму, запрошенная через метод .get_imdb().

Количество элементов в основном списке равняется количеству полученных с imdb данных о фильмах. Если какой-либо movieid в links.csv отсутствует, то результат его обработки в итоговый список добавлен не будет.

Давайте передадим в .get_imdb() список полей, которые мы хотим получить, например:

```python
%timeit links.get_imdb([193600, 2, 6, 2467, 903, 2176, 930, 2178], ['Title'])
links.get_imdb([193600, 2, 6, 2467, 903, 2176, 930, 2178], ['Title'])
```

Как мы увидели поле movieid добавилось автоматически- это нужно для того, чтобы результат не казался пустым если мы передали неправильное название поля:

```python
%timeit links.get_imdb([193600, 2, 6, 2467, 903, 2176, 930, 2178], ['title'])
links.get_imdb([193600, 2, 6, 2467, 903, 2176, 930, 2178], ['title'])
```

Из данного примера в процессе анализа MovieLens мы усволили, что запрашивать поля следует через заглавную букву, например, запросим еще бюджет и ссылку на страничку в интернете:

```python
%timeit links.get_imdb([193600, 2, 6, 2467, 903, 2176, 930, 2178], ['Title', 'Budget', 'URL'])
links.get_imdb([193600, 2, 6, 2467, 903, 2176, 930, 2178], ['Title', 'Budget', 'URL'])
```

Отлично! У нас всё отобразилось в нужном порядке. Кстати следует заметить, что бюджет фильма на imdb указан в национальной валюте :\) и в верхней строчке отображен фильм с бюджетом в 30 млрд Итальянских Лир, но об это позже...

Как видно, список списков отсортирован по убыванию movieid и это не случайно, потому как мы в "шапке" можем увидеть насколько глубока кроличья нора...

Чтобы продолжить, давате вызовем метод .get_imdb(), но без указания полей, чтобы получить всю доступную информацию для её дальнейшей обработки, ведь вызывая данный метод мы сохраняем полученные данные в атрибут нашего класса:

```python
%timeit links.get_imdb([193600, 2, 6, 2467, 903, 2176, 930, 2178])
links.get_imdb([193600, 2, 6, 2467, 903, 2176, 930, 2178])
```

Количество элементов основного списка == количество полученных данных о запрошенных фильмах (за исключением недоступных movieid) минус заголовок.

Результат парсинга мы также можем посмотреть в атрибуте .imdb_content

```python
%timeit links.imdb_content
links.imdb_content
```

Полученные данные, в зависимомости от запрошенных ранее полей, мы обрабатывали при помощи соответствующих методов. Каждый (из указанных ниже) метод возвращает словарь, где ключ- это имя САМОГО-САМОГО, а значение ключа- максимальная сумма достижений ключа, отсортированная по убыванию... Немного о доступных методах:

`.top_directors(top_n)` - отвечает за вывод словаря режиссеров, которые сняли больше всего фильмов из запрошенного списка

`.most_expensive(top_n)` - отвечает за вывод словаря фильмов, бюджет которых (в национальной валюте) был самым дорогим

`.most_profitable(top_n)` - отвечает за вывод словаря фильмов, которые оказались самыми прибыльными (разумеется, в баксах!)

`.longest(top_n)` - отвечает за вывод словаря самых продолжительных фильмов (в секундах)

`.top_cost_per_minute(top_n)` - отвечает за вывод словаря фильмов и во сколько им обошлась стоимость съемки одной минуты
___
*`.top_highest_grossing_directors(top_n)` - отвечает за вывод словаря режиссеров, фильмы которых окупились дороже всего
___
Методы `.get_top_sequence()` и `.isint()`- эти методы являются вспомогательными и они в непосредственной обраотке данных imdb не участвуют, но они оказывают существенное влияние на логику работы скрипта.
